```python
import matplotlib.pyplot as plt
from scipy import signal
image = x_train[0]/255
kernel = model.get_weights()[0]
maps1 = np.zeros((24,24,1,25))
for i in range(25):
    plt.figure(figsize=(10.0, 10.0))
    plt.subplot(121)
    plt.imshow(kernel[:,:,0,i],cmap='gray')
    plt.subplot(122)
    featmap = signal.correlate2d(np.reshape(image,(28,28)), kernel[:,:,0,i], boundary='symm', mode='valid')
    featmap = featmap + model.get_weights()[1][i]
    maps1[:,:,0,i] = featmap
    plt.imshow(featmap,cmap='gray')
    name = '25/'+str(i)+'.png'
    plt.savefig(name)
    name = '25/'+str(i)+'_kernel.csv'
    np.savetxt(name,kernel[:,:,0,i], delimiter=',')
    name = '25/'+str(i)+'_map.csv'
    np.savetxt(name,featmap, delimiter=',')
```

```python
plt.imshow(maps1[0],cmap='gray')
plt.plot()
```

```python
for i in range(12):
    for j in range(12):
        for k in range(25):
            if (maps1[i,j,0,k]<0):
                maps1[i,j,0,k] = 0
plt.imshow(maps1[:,:,0,0],cmap='gray')
plt.plot()

maps1_small = np.zeros((12,12,1,25))

for i in range(25):
    for k in range(12):
        for j in range(12):
            max = maps1[2*j,2*k,0,i]
            if (maps1[(2*j),(2*k),0,i] < maps1[(2*j),(2*k)+1,0,i]):
                max = maps1[(2*j),(2*k)+1,0,i]
            if (maps1[(2*j),(2*k)+1,0,i] < maps1[(2*j)+1,(2*k),0,i]):
                max = maps1[(2*j)+1,(2*k),0,i]
            if (maps1[(2*j)+1,(2*k),0,i] < maps1[(2*j)+1,(2*k)+1],0,i):
                max = maps1[(2*j)+1,(2*k)+1,0,i]
            maps1_small[j,k,0,i] = max
            

plt.imshow(maps1_small[:,:,0,0],cmap='gray')
```

```python
for i in range(25):
    plt.figure(figsize=(5.0, 5.0))
    
    plt.imshow(maps1_small[:,:,0,i],cmap='gray')
    name = '25_pool/'+str(i)+'.png'
    plt.savefig(name)
    name = '25_pool/'+str(i)+'_pooledMap.csv'
    np.savetxt(name,maps1_small[:,:,0,i], delimiter=',')
```
