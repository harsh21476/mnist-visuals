```python
%matplotlib inline
# %load kerass_conv.py

from __future__ import print_function
import matplotlib
matplotlib.use('Agg')

import sys
import keras
# from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
from sklearn.datasets import fetch_mldata
import matplotlib.pyplot as plt
import numpy as np
mnist = fetch_mldata('MNIST original')

batch_size = 1000
num_classes = 10
epochs = 100

# input image dimensions
img_rows, img_cols = 28, 28

# the data, shuffled and split between train and test sets
# (x_train, y_train), (x_test, y_test) = mnist.load_data()
X, y = mnist["data"], mnist["target"]
x_train, x_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
# print(x_train.shape,y_train.shape,x_test.shape)

```

```python
# print(x_train[0])
import sklearn
plt.figure(figsize=(10.0, 10.0))
plt.imshow(np.reshape(x_train[0],(28,28)),cmap='gray')
plt.show()
image = x_train[1]/255
plt.imshow(np.reshape(image,(28,28)),cmap='gray',vmin=0,vmax=1)
plt.show()
```

```python
from keras.models import load_model
model = load_model('my_model.h5')
```

```python
import matplotlib.pyplot as plt
from scipy import signal
image = x_train[0]/255
kernel = model.get_weights()[0]
maps1 = np.zeros((24,24,1,25))
for i in range(25):
    plt.figure(figsize=(10.0, 10.0))
    plt.subplot(121)
    plt.imshow(kernel[:,:,0,i],cmap='gray')
    plt.subplot(122)
    featmap = signal.correlate2d(np.reshape(image,(28,28)), kernel[:,:,0,i], boundary='symm', mode='valid')
    featmap = featmap + model.get_weights()[1][i]
    maps1[:,:,0,i] = featmap
    plt.imshow(featmap,cmap='gray')
    name = '25/'+str(i)+'.png'
    plt.savefig(name)
    name = '25/'+str(i)+'_kernel.csv'
    np.savetxt(name,kernel[:,:,0,i], delimiter=',')
    name = '25/'+str(i)+'_map.csv'
    np.savetxt(name,featmap, delimiter=',')
```

```python
plt.imshow(maps1[:,:,0,0],cmap='gray')
plt.plot()
```

```python
for i in range(12):
    for j in range(12):
        for k in range(25):
            if (maps1[i,j,0,k]<0):
                maps1[i,j,0,k] = 0
plt.imshow(maps1[:,:,0,0],cmap='gray')
plt.plot()

maps1_small = np.zeros((12,12,1,25))

for i in range(25):
    for k in range(12):
        for j in range(12):
            max = maps1[2*j,2*k,0,i]
            if (maps1[(2*j),(2*k),0,i] < maps1[(2*j),(2*k)+1,0,i]):
                max = maps1[(2*j),(2*k)+1,0,i]
            if (maps1[(2*j),(2*k)+1,0,i] < maps1[(2*j)+1,(2*k),0,i]):
                max = maps1[(2*j)+1,(2*k),0,i]
            if (maps1[(2*j)+1,(2*k),0,i] < maps1[(2*j)+1,(2*k)+1],0,i):
                max = maps1[(2*j)+1,(2*k)+1,0,i]
            maps1_small[j,k,0,i] = max
            

plt.imshow(maps1_small[:,:,0,0],cmap='gray')
```

```python
for i in range(25):
    plt.figure(figsize=(5.0, 5.0))
    
    plt.imshow(maps1_small[:,:,0,i],cmap='gray')
    name = '25_pool/'+str(i)+'.png'
    plt.savefig(name)
    name = '25_pool/'+str(i)+'_pooledMap.csv'
    np.savetxt(name,maps1_small[:,:,0,i], delimiter=',')
```

```python
kernel = model.get_weights()[2]
maps2 = np.zeros((8,8,1,50))

for i in range(50):
    plt.figure(figsize=(5.0, 5.0))
    plt.subplot(121)
    plt.imshow(kernel[:,:,0,i],cmap='gray')
    plt.subplot(122)
    for j in range(25):
        featmap = signal.correlate2d(maps1_small[:,:,0,j], model.get_weights()[2][:,:,j,i], boundary='symm', mode='valid')
        featmap = featmap + model.get_weights()[3][i]
        maps2[:,:,0,i] = featmap
    plt.imshow(featmap,cmap='gray')
    name = '50/'+str(i)+'.png'
    plt.savefig(name)
    name = '50/'+str(i)+'_kernel.csv'
    np.savetxt(name,kernel[:,:,0,i], delimiter=',')
    name = '50/'+str(i)+'_map.csv'
    np.savetxt(name,featmap, delimiter=',')
```

```python
for i in range(8):
    for j in range(8):
        for k in range(50):
            if (maps2[i,j,0,k]<0):
                maps2[i,j,0,k] = 0
plt.imshow(maps2[:,:,0,0],cmap='gray')
plt.plot()

maps2_small = np.zeros((4,4,1,50))

for i in range(50):
    for k in range(4):
        for j in range(4):
            max = maps2[2*j,2*k,0,i]
            if (maps2[(2*j),(2*k),0,i] < maps2[(2*j),(2*k)+1,0,i]):
                max = maps2[(2*j),(2*k)+1,0,i]
            if (maps2[(2*j),(2*k)+1,0,i] < maps2[(2*j)+1,(2*k),0,i]):
                max = maps2[(2*j)+1,(2*k),0,i]
            if (maps2[(2*j)+1,(2*k),0,i] < maps2[(2*j)+1,(2*k)+1],0,i):
                max = maps2[(2*j)+1,(2*k)+1,0,i]
            maps2_small[j,k,0,i] = max
            

plt.imshow(maps2_small[:,:,0,0],cmap='gray')

for i in range(50):
    plt.figure(figsize=(5.0, 5.0))
    
    plt.imshow(maps2_small[:,:,0,i],cmap='gray')
    name = '50_pool/'+str(i)+'.png'
    plt.savefig(name)
    name = '50_pool/'+str(i)+'_pooledMap.csv'
    np.savetxt(name,maps2_small[:,:,0,i], delimiter=',')
```
